# 1.2.4 Funkce

```python
import math
def measure(r:float, dim:int=3, shape:str='cube') -> float:
    """
    Print and returns a measure (length/area/volume) of a simple shape.
    
    param r: side of cube or radius of a sphere.
    param dim: dimension of the shape
    param shape: shape type
    return : the measure value
    """
    
    if shape == 'cube':
        m = r**dim
    elif shape == 'sphere':
        m = math.pi * r**dim
    else:
        raise ValueError("Unknown shape '{}'.".format(shape))
    print("Míra: ", m)
    return m

measure(1)
measure(2)
measure(2, 2)
measure(1, shape='sphere')
measure(1, shape='box')
```

# 2.2 Elipsy

```python
import numpy as np
import matplotlib.pyplot as plt

def rot_mat(angle):
    return np.array([[np.cos(angle), -np.sin(angle)], [np.sin(angle), np.cos(angle)]])

a, b = 3, 1
n = 17
N = 100

fig, ax = plt.subplots(1, 1, figsize=(10,10))
for i in range(n):
    t = np.linspace(0, 2 * np.pi, N)
    XY = np.stack((a * np.cos(t), b * np.sin(t)))
    XY = rot_mat(np.pi * i / n) @ XY
    ax.scatter(*XY, c=t)
```
